#carregando o modelo Phi-3 Mini Instruct, baixando e preparando o tokenizer, configurando o modelo para utilizar FP16, definindo automaticamente se a execução será feita na GPU ou na CPU e criando um pipeline de geração de texto que permite enviar um prompt ao modelo e receber uma resposta de forma simplificada.

```python
MODEL_ID = "microsoft/phi-3-mini-4k-instruct"
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch


tokenizer = AutoTokenizer.from_pretrained(MODEL_ID)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_ID,
    device_map="auto",
    torch_dtype=torch.float16,
)

generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    device_map="auto",
    max_new_tokens=200
)
```

# Salvando modelo

```python
model.save_pretrained("careerup-model-fp16", safe_serialization=True)
tokenizer.save_pretrained("careerup-model-fp16")
```

# Testando o modelo

```python
def gerar_prompt(cargo, h1, h2, h3):
    return f"""
Você é uma IA especialista em análise de carreira e recomendação profissional.

Gere uma recomendação personalizada para um usuário com os seguintes dados:

Cargo atual: {cargo}
Habilidade principal: {h1}
Habilidade secundária: {h2}
Habilidade terciária: {h3}

Sua resposta deve:
- Ser objetiva e clara.
- Conter sugestões de evolução profissional.
- Incluir áreas, cursos ou frameworks relevantes.
- Indicar possíveis vagas que combinem com o perfil.
- Ter no máximo 8 linhas.

Agora gere a recomendação:
"""

cargo = "Desenvolvedor Front-end"
h1 = "React"
h2 = "JavaScript"
h3 = "UX Design"

prompt = gerar_prompt(cargo, h1, h2, h3)

resposta = generator(prompt)[0]["generated_text"]

print("==== RESPOSTA DA IA ====\n")
print(resposta)
```
